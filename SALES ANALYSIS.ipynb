#### Products that are often sold together

```python
df = clean_df[clean_df['Order ID'].duplicated(keep= False)]
df.head(20)
```

```python
df['grouped'] = df.groupby('Order ID')['Product'].transform( lambda x: ','.join(x))
```

```python
df.head()
```

```python
df = df[['Order ID','grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df['grouped']: 
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))


```

```python
gd = count.most_common()
gd = pd.DataFrame( gd )
gd.head().reset_index()
```

```python
gd[0] = gd[0].apply(lambda x: list(x))
print(gd)
gd.reset_index()
gd.head()
```

```python
gd[0]= gd[0].astype('str')
gd[['object1','object2']]= gd[0].str.strip('[]').str.split(',', expand=True)
gd.drop([0], inplace = True , axis = 1)
gd
```

```python
gd['freq'] = gd[1]
```

```python
gd.head()
```

```python
gd.dropna(inplace=True)
gd.fillna(0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(gd[['object1','freq']], gd['object2'])
```

```python
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train['object1'])
X_train_vec = pd.DataFrame(X_train_vec.toarray(), columns=vectorizer.get_feature_names_out())
X_train_vec['freq'] = X_train['freq']
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
print(X_test_vec)
```

```python
X_test_vec = vectorizer.transform(X_test['object1'])
X_test_vec = pd.DataFrame(X_test_vec.toarray(), columns=vectorizer.get_feature_names_out())
print(X_test_vec)
```

```python
input_object = 'Google Phone'
input_vec = vectorizer.transform([input_object])
predicted_object = model.predict(input_vec)
```
